## Scrapped *Flipkart* Product Reviews 
- Name of **Reviewer**
- Review in **Numeric** 
- Review **Title**
- Review **Description**
- Review **DATE**

```python
reviews = beautified_html.find_all("div", {"class":"_16PBlm"})
reviews.pop() # Removing Last Element Because It's not a review
```

```python
for i in reviews: 
    reviewerName = i.find('p', {"class":"_2mcZGG"}).text;
    title = i.find("div", {"class": "row"}).p.text;
    date = i.div.div.find_all("p", {"class": "_2sc7ZR"})[1].text;    
    ratingNum = i.find("div", {"class": "_3LWZlK"}).text;
    desc = i.find("div", {"class": "t-ZTKy"}).text
    
    print(reviewerName,"\n" ,  ratingNum, "",'\033[1m',  title  ,'\033[0m')
    print(desc , "\n-" , date , "\n")
```

### Storing The Data IN CSV FORMAT
Using Panadas

```python
import pandas as pd

nameList, titleList, dateList, ratingNumList, descList = [], [], [], [], []

for i in reviews: 
    reviewerName = i.find('p', {"class":"_2mcZGG"}).text[16:];
    title = i.find("div", {"class": "row"}).p.text;
    date = i.div.div.find_all("p", {"class": "_2sc7ZR"})[1].text;    
    ratingNum = i.find("div", {"class": "_3LWZlK"}).text;
    desc = i.find("div", {"class": "t-ZTKy"}).text
    
    nameList.append(reviewerName)
    titleList.append(title)
    dateList.append(date)
    ratingNumList.append(ratingNum)
    descList.append(desc)
    
#     print(reviewerName,"\n" ,  ratingNum, "",'\033[1m',  title  ,'\033[0m')
#     print(desc , "\n-" , date , "\n")
    
dict_data = {
    "name":nameList, 
    "title": titleList, 
    "date": dateList, 
    "rating_num": ratingNumList, 
    "long_rating": descList
}    

df = pd.DataFrame(dict_data);
df
```
